```python
# Lesson 1: Potential Flows
import math
import numpy as np
from matplotlib import pyplot as py
```

```python
N = 50
x_start, x_end = -2.0, 2.0
y_start, y_end = -1.0, 1.0
x = np.linspace(x_start,x_end,N)
y = np.linspace(y_start,y_end,N)

#print('x = ', x)
#print('y = ', y)
X,Y = np.meshgrid(x,y)
```

```python
%matplotlib inline

width = 10.0 
height = (y_end - y_start)/(x_end-x_start)*width
py.figure(figsize=(width,height))
py.xlabel('x',fontsize=18)
py.ylabel('y',fontsize=18)
py.xlim(x_start,x_end)
py.ylim(y_start,y_end)
py.scatter(X,Y, s=5, color='#CD2310', marker='o')
```

```python
strength_source = 5.0                      # source strength
x_source, y_source = -1.0, 0.0             # location of the source

# compute the velocity field on the mesh grid
u_source = (strength_source/(2 * math.pi)) *(X - x_source) / ((X - x_source)**2 + (Y - y_source)**2)
v_source = (strength_source/(2 * math.pi)) *(Y - y_source) / ((X - x_source)**2 + (Y - y_source)**2)
```

```python
# plot the streamlines
width = 10.0
height = (y_end - y_start) / (x_end - x_start) * width
py.figure(figsize=(width, height))
py.xlabel('x', fontsize=16)
py.ylabel('y', fontsize=16)
py.xlim(x_start, x_end)
py.ylim(y_start, y_end)
py.streamplot(X, Y, u_source, v_source,
                  density=2, linewidth=1, arrowsize=2, arrowstyle='->')
py.scatter(x_source, y_source,
               color='#CD3510', s=80, marker='o');
```

```python
V_s = np.sqrt(u_source**2 + v_source**2)
r = 0.1
A = math.pi * r**2
mass_flux = V_s*A
```

```python
## Sink
strength_sink = -5.0                      # source strength
x_sink, y_sink = 1.0, 0.0             # location of the source

# compute the velocity field on the mesh grid
u_sink = (strength_sink/(2 * math.pi)) *(X - x_sink) / ((X - x_sink)**2 + (Y - y_sink)**2)
v_sink = (strength_sink/(2 * math.pi)) *(Y - y_sink) / ((X - x_sink)**2 + (Y - y_sink)**2)
```

```python
# plot the streamlines
width = 10.0
height = (y_end - y_start) / (x_end - x_start) * width
py.figure(figsize=(width, height))
py.xlabel('x', fontsize=16)
py.ylabel('y', fontsize=16)
py.xlim(x_start, x_end)
py.ylim(y_start, y_end)
py.streamplot(X, Y, u_sink, v_sink,
                  density=2, linewidth=1, arrowsize=2, arrowstyle='->')
py.scatter(x_sink, y_sink,
               color='#CD3510', s=80, marker='o');
```

## Source-sink Pair

```python
strength_source = 10.0                      # source strength
x_source, y_source = -0.7, 0.2             # location of the source

# compute the velocity field on the mesh grid
u_source = (strength_source/(2 * math.pi)) *(X - x_source) / ((X - x_source)**2 + (Y - y_source)**2)
v_source = (strength_source/(2 * math.pi)) *(Y - y_source) / ((X - x_source)**2 + (Y - y_source)**2)


## Sink
strength_sink = -20.0                      # source strength
x_sink, y_sink = 1.0, -0.5             # location of the source

# compute the velocity field on the mesh grid
u_sink = (strength_sink/(2 * math.pi)) *(X - x_sink) / ((X - x_sink)**2 + (Y - y_sink)**2)
v_sink = (strength_sink/(2 * math.pi)) *(Y - y_sink) / ((X - x_sink)**2 + (Y - y_sink)**2)


## velocity fields values
u_pair = u_sink + u_source
v_pair = v_sink + v_source

## Plot line
width = 10.0
height = (y_end - y_start) / (x_end - x_start) * width
py.figure(figsize=(width, height))
py.xlabel('x', fontsize=16)
py.ylabel('y', fontsize=16)
py.xlim(x_start, x_end)
py.ylim(y_start, y_end)
py.streamplot(X, Y, u_pair, v_pair,
                  density=2, linewidth=1, arrowsize=2, arrowstyle='->')
py.scatter([x_source,x_sink],[y_source, y_sink],
               color='#CD3510', s=80, marker='o');
```

```python
# Contour plot
## Plot line
U_sink = np.sqrt(u_sink**2+v_sink**2)
U_source = np.sqrt(u_source**2+v_source**2)
U_pair = np.sqrt(u_pair**2+v_pair**2)
width = 10.0
height = (y_end - y_start) / (x_end - x_start) * width
py.figure(figsize=(width, height))
py.xlabel('x', fontsize=16)
py.ylabel('y', fontsize=16)
py.xlim(x_start, x_end)
py.ylim(y_start, y_end)
py.contour(X,Y,U_pair,cmap='jet',extend='max');
```
